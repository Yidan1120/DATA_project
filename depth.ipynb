```python
import pandas as pd
import os
```

```python
PATH = 'individual_book_train/'
files = [PATH+i for i in os.listdir('individual_book_train') if '.csv' in i]
file_names = [i[:-4] for i in os.listdir('individual_book_train') if '.csv' in i]
```

```python
doms = []
spreads = []
for i in range(len(files)):
    print(i/len(files))
    file = files[i]
    stock = pd.read_csv(file)
    stock["DOM"] = stock['bid_price1']*stock['bid_size1']+stock['bid_price2']*stock['bid_size2']+stock['ask_price1']*stock['ask_size1']+stock['ask_price2']*stock['ask_size2']
    stock["spread"] = stock['ask_price1']/stock['bid_price1'] - 1
    dom = stock["DOM"].to_numpy().mean()
    spread = stock["spread"].to_numpy().mean()
    doms.append(dom)
    spreads.append(spread)
```

```python
df = pd.DataFrame({"DOM": doms,"spreads": spreads},index=file_names)
```

```python
df.to_csv("depth_spread.csv")
```
